Q1. What is Flask Framework? What are the advantages of Flask Framework?


Flask is a micro web framework for Python, designed to be lightweight, flexible, and easy to use. It is classified as a "micro" framework because it does not include a built-in toolset for database abstraction, form validation, or other components typically found in larger web frameworks. Instead, Flask focuses on providing the core functionalities needed to build web applications, leaving the choice of additional libraries or extensions to the developers.

Advantages:

Lightweight and simple

Flexible and allows using preferred components

Easy to get started with minimal code

Extensible through Flask extensions

Good support for building RESTful APIs

Well-documented with detailed explanations and examples

Large and active community support

Seamless integration with Python libraries and tools.

Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask:

In Flask, app routing refers to the process of associating specific URL paths (routes) with Python functions or view handlers. When a user makes a request to a particular URL, Flask's routing system determines which function should handle that request based on the URL path. This allows you to map different URLs to different functions and, consequently, different pages or actions in your web application.

**URL Mapping:** App routes allow you to map specific URLs to corresponding functions or view handlers. This organization makes your code modular and easy to maintain, as each function handles a specific part of your application.

**Separation of Concerns:** By using app routes, you can separate different parts of your application's logic into distinct functions. For example, you can have one function to handle user authentication, another for serving content, and yet another for processing form submissions. This promotes better code organization and makes it easier to debug and maintain.

**Clean and Readable Code:** Routing in Flask enables you to define URL patterns in a clean and straightforward manner. This clarity makes it easy to understand the flow of your application and aids other developers working on the project.

**RESTful APIs:** App routes are particularly useful when building RESTful APIs. You can map different routes to different HTTP methods (e.g., GET, POST, PUT, DELETE) to handle various operations on the API resources.

**Dynamic URLs:** Flask allows you to define dynamic routes by using variable parts in the URL. These variables can be captured and passed as arguments to your functions, allowing you to create dynamic and data-driven web applications.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    details = {
        'Company Name': 'ABC Corporation',
        'Location': 'India',
        'Contact Detail': '999-999-9999'
    }
    response = "<br>".join([f"{key}: {value}" for key, value in details.items()])
    return response

if __name__ == '__main__':
    app.run()

```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.


In Flask, the function used for URL building is url_for(). The url_for() function is a powerful utility provided by Flask that helps generate URLs for the specified routes or functions in your application. It abstracts the URL generation process, making it easier to manage and maintain URLs, especially when your routes change.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def index():
    # Using url_for() to generate URLs for the 'welcome' and 'company_details' routes
    welcome_url = url_for('welcome')
    company_details_url = url_for('company_details')
    
    return f'Click <a href="{welcome_url}">here</a> to visit the Welcome page.<br>' \
           f'Click <a href="{company_details_url}">here</a> to visit the Company Details page.'

@app.route('/company')
def company_details():
    return 'Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999'

if __name__ == '__main__':
    app.run()
```
